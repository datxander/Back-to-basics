Dictionaries
A dictionary is a collection of unordered, modifiable(mutable) paired (key: value) data type.

Dictionary_applications.jpg


```python
# Creating a dictionary

empty_dict = {}
starter_dict = {'key1' : 'value1','key2' : 'value2', 'key3' : 'value3', 'key4' : 'value4'}
```

```python
# Example : Note that a value could be any data types:string, boolean, list, tuple, set or a dictionary.

person = {
    'first_name':'Dante',
    'last_name':'Allegri',
    'age':41,
    'country':'India',
    'is_marred':True,
    'skills':['SQL', 'Python', 'Tableau', 'Power BI', 'Analytics'],
    'address':{
        'street':'Space street',
        'zipcode':'02210'
    }
    }
```

```python
# We can pull up a dictionary item through its key name

print(person['age'])
```

```python
# This method will throw up an error if the specified key does not exist. To avoid this, use the get() method for the key

print(person.get('first_name'))
```

```python
# Add / append items to the dictionary

person['gender'] = 'Male'
person['skills'].append('Team leadership')
print(person)
```

```python
# Modify dictionary items by directly updating the key value pair

person['first_name'] = 'Inferno'
print(person)
```

```python
# Check if a key exists in the dictionary
print('hobbies' in person)
```

```python
# Removing Key - Value Pairs from a Dictionary with pop() and del

person.pop('country')   # Will remove the specified key
del person['age']      # Will also remove the specified key
person.popitem()      # Will remove the last item from the dict
print(person)
```

**When should a dictionary be converted to a list? **



1.   Iteration and Looping: If you need to iterate over key-value pairs in a specific order.

2.   Sorting: To sort the dictionary based on keys or values.

3.   Data Transformation: When you need to pass data to functions that expect lists.

4. Data Export: For exporting data to formats like CSV or JSON, which often require list structures.

5. Zipping/Unpacking: To pair keys and values or perform operations on them individually.






```python
# How to convert the dictionary into a list of items?

dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
print(dct.items())
```

```python
# Clear / Delete the dictionary

print(dct.clear())
del dct
```

```python
# Copy the dictionary

dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
dct2 = dct.copy()
```

```python
# Retrieve dictionary keys as a list
keys = dct.keys()
print(keys)

values = dct.values()
print(values)
```
