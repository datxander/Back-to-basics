A tuple is a collection of different data types which is ordered and unchangeable (immutable). Tuples are written with round brackets, (). Once a tuple is created, we cannot change its values. We cannot use add, insert, remove methods in a tuple because it is not modifiable (mutable). Unlike list, tuple has few methods.

In essence, lists are great for collections of items that might change, while tuples are ideal for collections of items that should remain constant. Choosing between them depends on whether you need to modify the data or keep it fixed.

# Lists vs tuples

**List properties:**

Mutable: Lists can be changed after their creation, making them suitable for tasks where data needs to be updated frequently.

Ordered Collection: Lists maintain the order of elements, so they are ideal for storing sequences of items.

Flexible Size: Lists can grow or shrink in size dynamically.

**Real-life applications:**

Todo Lists: Managing tasks that need to be updated, added, or removed.

Shopping Cart: In e-commerce applications, where users can add or remove items.

Data Analysis: Storing data sets that can be manipulated and analyzed.

Gaming: Keeping track of players or game states that can change frequently.



**Tuples properties**

Immutable: Tuples cannot be changed once created, making them suitable for tasks where data integrity is crucial.

Ordered Collection: Tuples also maintain the order of elements.

Hashable: Tuples can be used as keys in dictionaries because they are immutable.

**Real-life applications:**

GPS Coordinates: Storing fixed pairs of latitude and longitude values.

Configuration Settings: Storing settings or preferences that should not be altered.

Database Records: Representing rows in a database where each tuple corresponds to a fixed record.

Function Arguments: Using tuples to return multiple values from a function securely.

```python
# Tuple methods

# Creating an empty tuple

empty_tuple = ()        # Regular syntax

empty_tuple = tuple()   # Using the tuple constructor
```

```python
# Initialize a tuple

tpl = ('key1','key2', 'key3')
fruits = ('banana', 'apple', 'orange')
```

```python
# Tuple length

len(tpl)
```

```python
# Indexing in a tuple

# Positive indexing still works

fruits = ('banana', 'orange', 'mango', 'lemon')
first_fruit = fruits[0]
last_index = len(fruits) - 1
last_fruit = fruits[last_index]

# So does negative indexing
last_fruit = fruits[-1]
```

```python
# Slicing also works the same as with lists


all_fruits = fruits[0:]
fruit_slices = fruits[:-1]
print(fruit_slices )
```

**Changing Tuples to Lists**

We can change tuples to lists and lists to tuples. A tuple is immutable, so if we want to modify a tuple we should change it to a list.

```python
# Conversion code
fruits = list(fruits) # Tuple to list conversion
fruits.append('kiwi')
print(fruits)

fruits = tuple(fruits) # List to tuple conversion
print(fruits)
```

```python
# How to search for an item in a tuple - This returns a boolean result

'kiwi' in fruits
print('coconut' in fruits)
```

```python
# Joining / Appending  whole tuples works in the same way as lists. Remember that we can't add individual items directly to a tuple

tup1 = (1,2,3)
tup2 = (4,5,6)
tup3 = tup1 + tup2
print(tup3)
```

```python
# We also can't remove an item from a tuple, but the whole tuple could be deleted

del tup1
del tup2
del tup3
```
