**Introduction**

A set is a collection of unordered and un-indexed distinct elements, the same as we'd learnt in our school Math textbooks. In Python the set is used to store unique items, and it is possible to find the union, intersection, difference, symmetric difference, subset, super set and disjoint set among sets

```python
st = set()                                             # Create an empty set
fruits = {'banana', 'orange', 'mango', 'lemon'}        # Initialize the set
```

```python
len(fruits)   # Length of set
```

```python
# Check whether a set contains an item with the IN operator

print('mango' in fruits)

```

*Once a set is created we cannot change any items, but we can add or remove items.*

```python
fruits.add('kiwi')                                     # Using add() - Allows 1 item at a time. Duplicate items are not added
fruits.update(['strawberry', 'guava', 'coconut'])    # Using update() - Allows multiple items to be added through a list argument
print (fruits)


num1 = {'item1','item2'}
num2 = {'item3', 'item4'}
num1.update(num2)                                     # Updates take in other lists too
print(num1)                                           # NOTE : The order appears random
```

```python
# Remove items from a set

num1.remove('item4')        # With remove(). If an item is not present in the set, remove() will throw an error
num1.discard('item3')       # With discard(). If an item is not present in the set, discard() will NOT throw an error
print(num1)

fruits.pop()                # pop() removes a random item from the list and returns the removed item
removed_item = fruits.pop() # If we're interested in capturing the popped item, we can assign it to a var
print(removed_item)
```

**How are sets different from tuples?**



*   Sets are best for managing collections of unique items and performing operations on sets of data. This makes them extremely useful for situations where you need to maintain a collection of items without duplicates.

*   Tuples are ideal for representing fixed collections of related items, especially when order and immutability are important. They are used in scenarios where a fixed collection of items is needed and where each element has a specific meaning or order.

```python
# Clearing the set vs Deleting it

num1.clear()
print(num1)

fruits = {'banana', 'orange', 'mango', 'lemon'}
del fruits
```

```python
# Conversions

  # Convert a list to a set - This is useful for deduping a list quickly
lst = [1,2,2,2,3,3,4,5,6,7,7,8]
st = set(lst)
print(st) # The set may be unordered, as that is a property of sets

  # Convert the set back to a list
lst = list(st)
print(lst)

```

```python
# Join sets

  # Union() returns a new set
num1 = {1,2,3,4}
num2 = {5,6,7,8}
num3 = num1.union(num2)
print (num3)

  # Update() adds to the original set
num1.update(num2)
print(num1)
```

```python
# Finding intersection items

num1 = {1,2,3,4,5}
num2 = {1,2,3}
num1.intersection(num2)
```

```python
# Check for Subsets and Super Sets

num2.issubset(num1)
num1.issuperset(num2)
```

```python
# Differences between sets

num1.difference(num2) # Difference between 2 sets
num1.symmetric_difference(num2) #  (A\B)  (B\A)

```

```python
# Disjoint sets check

even_numbers = {2,4,6,8}
odd_numbers = {3,5,7,9}
even_numbers.isdisjoint(odd_numbers)
```
